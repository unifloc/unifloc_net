```python
import numpy as np
import matplotlib as mpl
from matplotlib import pylab as plt
```

```python
d_mm = 25
p_in_atma = 60
p_out_atma = 50
gamma_g = 0.9
t_c = 80
q_gas, q_gas_dll, q_gas_v, q_gas_vdll, p_gas_v, p_gas_vdll, p_gas_, p_gas_dll = [], [], [], [], [], [], [], []
d_vkr_mm = 50
p_open, p_open_dll, dp, dp_dll = [], [], [], []
d_0 = 8
d_n = 36
dd = 2 
d_mm_ = np.arange(d_0, d_n, dd) #
for d in d_mm_:
    q_ = UniflocVBA.GLV_q_gas_sm3day(d, p_in_atma, p_out_atma, gamma_g, t_c)
    q_gas.append(q_[0][0])
    q_dll = u7_excel.u7_Excel_functions_GL.GLV_q_gas_sm3day(d, p_in_atma, p_out_atma, gamma_g, t_c)
    q_gas_dll.append(q_dll[0])
    q_v = UniflocVBA.GLV_q_gas_vkr_sm3day(d, d_vkr_mm, p_in_atma, p_out_atma, gamma_g, t_c)
    q_gas_v.append(q_v[0][0])
    q_vdll = u7_excel.u7_Excel_functions_GL.GLV_q_gas_vkr_sm3day(d, d_vkr_mm, p_in_atma, p_out_atma, gamma_g, t_c)
    q_gas_vdll.append(q_vdll[0])
    p_v = UniflocVBA.GLV_p_vkr_atma(d, d_vkr_mm, p_in_atma, 50, gamma_g, t_c)
    p_gas_v.append(p_v[0][0])
    p_vdll = u7_excel.u7_Excel_functions_GL.GLV_p_vkr_atma(d, d_vkr_mm, p_in_atma, 50, gamma_g, t_c)
    p_gas_vdll.append(p_vdll[0])
    p_v = UniflocVBA.GLV_p_atma(d, p_in_atma, 50, gamma_g, t_c)
    p_gas_.append(p_v[0][0])
    p_vdll = u7_excel.u7_Excel_functions_GL.GLV_p_atma(d, p_in_atma, 50, gamma_g, t_c)
    p_gas_dll.append(p_vdll[0])
    p_v = UniflocVBA.GLV_IPO_p_open(p_in_atma, p_out_atma, t_c, 0, d)
    p_open.append(p_v)
    p_vdll = u7_excel.u7_Excel_functions_GL.GLV_IPO_p_open(p_in_atma, p_out_atma, t_c, 0, d)
    p_open_dll.append(p_vdll)
    #p_v = UniflocVBA.GL_dPgasPipe_atmg(100, 50, 80, 125, 73, 0.9, d*10**3)
    #dp.append(p_v)
    p_vdll = u7_excel.u7_Excel_functions_GL.GL_dPgasPipe_atmg(100, 50, 80, 125, 73, 0.9, d*10**3)
    dp_dll.append(p_vdll)
```

```python
mpl.rcParams['font.family'] = 'fantasy'
mpl.rcParams['font.fantasy'] = 'Times New Roman'
ys = [ q_gas, q_gas_dll]

for j in range(2):
        plt.grid(True)
        plt.title('Расход газа через гл клапан от диаметра', color='black', family='fantasy')
        plt.ylabel('Расход газа через клапан', color='black', family='fantasy')
        plt.xlabel('Диаметр клапана', color='black', family='fantasy')
        plt.plot(d_mm_, ys[j], linewidth=3)
plt.legend(('vba', 'vb.net')) 
plt.show()
```
